```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from pylab import rcParams
np.random.seed(1)
```

```python
def is_pos_def(X, nn = False):
    return np.all(np.linalg.eigvals(X) > 0) if nn == False else np.all(np.linalg.eigvals(X) >= 0)
```

```python
N = 101
```

```python
SIGMA = np.empty((3,3))
for i in range(3):
    for j in range(i,3):
        sigma = np.random.chisquare(df = N)/N
        SIGMA[j,i] = SIGMA[i,j] = np.random.chisquare(df = N)/N if i == j else np.random.randn()*sigma
SIGMA = SIGMA @ SIGMA.T

print(f'SIGMA > 0 : {is_pos_def(SIGMA, nn = True)}')
C = np.linalg.cholesky(SIGMA)
mu = np.reshape(np.random.uniform(-1,1, size = 3),(1,3))
```

```python
def MSE(y,y_hat):
    return np.mean((y - y_hat)**2)
def MAPE(y,y_hat, eps = 10**(-3)):
    return np.mean(np.abs((y - y_hat)/y))
def resids(y, y_hat, p=1):
    return (y-y_hat)**p
```

```python
def ridge_reg(y,X, lambd = 0):
    w = np.linalg.inv(X.T@X + lambd*np.eye(X.shape[1]))@X.T@y
    w = np.reshape(w, (X.shape[1],1))
    return w
```

## Pearson's chi-squared test

```python
from scipy.stats import chi2
```

```python
def chi_square(X, k, cdf, unknown_params = 0):
    l = len(X)
    a, b = np.min(X), np.max(X)
    h = (b-a)/k
    n, p = np.zeros((k,), dtype=int), np.zeros((k,), dtype=float)
    for i in range(k):
        ai, bi = a + h*i, a + h*(i+1)
        n[i] = np.sum((ai < X) & (X <= bi))
        p[i] = cdf(bi) - cdf(ai)# if cdf is not None else 0
    E = l*p
    stat = np.sum((n - E)**2/E)
    p_value = 1 - chi2.cdf(stat, df=k-1-unknown_params)
    return stat, p_value
```
